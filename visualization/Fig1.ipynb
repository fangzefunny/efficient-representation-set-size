```python
beh_data = get_beh_data('setsize-collins12', 'human')
beh_data['block_type'] = beh_data['block_type'].astype(int)
fig, ax = plt.subplots(1, 1, figsize=(3, 2.5))
sns.lineplot(x='repetitions', y='r', data=beh_data.query('repetitions<10'), 
             errorbar=('ci', 95), #err_style='bars', err_kws={'capsize': 3},
             hue='block_type', 
             ax=ax)
ax.set_xlabel('Repetitions')
ax.set_ylabel('Accuracy')
ax.legend(title='Set size')
ax.set_yticks([0.2, 0.4, 0.6, 0.8, 1])
ax.set_xticks([1, 3, 5, 7, 9])
ax.spines['left'].set_position(('axes',-0.04))
plt.savefig('../figures/Fig1/setsize.pdf', dpi=300)
plt.show()
```
