```python
def get_beh_data(data_set, model, method='mle', cross_valid=True):
    if model=='human':
        fname = f'../data/{data_set}-human.csv'
    else:
        cv = '-cv' if cross_valid else ''
        fname = f'../simulations/{data_set}{cv}/{model}/sim-{method}.csv'
    return pd.read_csv(fname)
```

```python
data = get_beh_data('collins_12', 'human')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
sns.lineplot(x='repetitions', y='r', data=data.query('repetitions<10'), 
             errorbar=('ci', 95), #err_style='bars', err_kws={'capsize': 3},
             hue='set_size', 
             ax=ax)
plt.show()
```
